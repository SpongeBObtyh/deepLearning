```python
import torch

x = torch.arange(4.0)
x.requires_grad_(True) 
```

```python
x.grad
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad 
```

```python
x.grad.zero_()
y = x * x
y.sum().backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad == u
```

```python
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```
