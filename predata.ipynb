```python
import os
print("当前工作目录是：", os.getcwd())
os.makedirs(os.path.join(os.getcwd(), 'data'), exist_ok=True)
data_file = os.path.join(os.getcwd(), 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')  # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

```python
import pandas as pd
data = pd.read_csv(data_file)
print(data)
```

```python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
numeric_inputs = inputs.iloc[:, 0:1] 
numeric_inputs = numeric_inputs.fillna(numeric_inputs.mean())

categorical_inputs = inputs.iloc[:, 1:2]
categorical_inputs = pd.get_dummies(categorical_inputs, dummy_na=True,dtype=int)

inputs = pd.concat([numeric_inputs, categorical_inputs], axis=1)
print(inputs)
type(inputs)
```
